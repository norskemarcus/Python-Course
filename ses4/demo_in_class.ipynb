## Ex 2: Clothes List Comprehension

From 2 lists, using a list comprehension, create a list containing:

```python
colors = ['Black', 'White']
sizes = ['s', 'm', 'l', 'xl']

# outputtet er en tuple (color,size)
result = [ (color, size) for color in colors for size in sizes ]

print(result)
# [('Black', 's'), ('Black', 'm'), ('Black', 'l'), ('Black', 'xl'), ('White', 's'), ('White', 'm'), ('White', 'l'), ('White', 'xl')]
```

```python
## If the tuple pair is in the following list, it should not be added to the comprehension generated list.

colors = ['Black', 'White']
sizes = ['s', 'm', 'l', 'xl']
sold_out = {('Black', 'm'), ('White', 's')}  # Use curly braces for set, not curly and square braces

not_sold_out = [(color, size) for color in colors for size in sizes if (color, size) not in sold_out]

print(not_sold_out)
```

```python
not_sold_out = [(color, size) if (color, size) not in sold_out else 'sold_out' for color in colors for size in sizes ]
```

## Ex. 3 List Comprehension exercises

```python
# Create a list of even numbers from 0 to 20.

evenNumbers = [ x for x in range(1,20+2) if x % 2 == 0 ]
print('Even numbers:', evenNumbers)

# Create a list of squares of numbers from 1 to 10 - husk da at inkl 10!
squares = [x * x for x in range(1,10+1)]
print('Squares:', squares)

# Create a list of all the vowels in a given string.
word = "Hello world"
vowels = [ l for l in word if l.lower() in 'aeiou']
print('Vowels: ', vowels)

# Create a list of common elements in two given lists.
list1 = [1,2,3,4]
list2 = [3,4,5,6]
common_elements = [e for e in list1 if e in list2]
print(common_elements)

# Create a list of words from a given string that have more than 4 letters: HUSK SPLIT!!!!
long_string = "This is a long sentence with more than four letters"
list_of_words = [word for word in long_string.split() if len(word) > 4]
print('List of words with more than 4 letters:', list_of_words)
```
